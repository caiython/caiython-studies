```python
class Hangman:
  #Método Construtor
  def __init__(self,word):
    self.word = word
    self.char_counter_word = len(word)
    self.char_counter_player = 0
    self.status = 0
    self.hidden_word = word
    self.charlist_input = []
    self.charlist_correctinput = []
    self.charlist_incorrectinput = []
  #Método para adivinhar a letra
  def guess(self,letra):
    self.letra = letra
    self.charlist_input.append(self.letra)
    if self.letra in self.word:
      self.charlist_correctinput.append(self.letra)
    else:
      self.charlist_incorrectinput.append(self.letra)
      self.status += 1
  #Método para verificar se o jogo terminou
  def hangman_over(self):
    self.char_counter_player = 0
    for i in self.word:
      if i in self.charlist_input:
        self.char_counter_player += 1
    if self.status == 6:
      return True
    elif self.char_counter_player == self.char_counter_word:
      return True
    else:
      return False
  #Método para verificar se o jogador venceu
  def hangman_won(self):
    if self.char_counter_player == self.char_counter_word:
      return True
    else:
      return False
  #Método para não mostrar a letra no board
  def hide_word(self):
    print("\nPalavra: ", end='')
    for i in self.word:
      if i in self.charlist_input:
        print(i,end='')
      else:
        print('_',end='')
  #Método para mostrar as letras incorretas
  def wrong_chars(self):
    print("\n\nLetras erradas: ", end='')
    for i in self.charlist_incorrectinput:
      print(i,end=' ')
  #Método para mostrar as letras corretas
  def correct_chars(self):
    print("\n\nLetras corretas: ", end='')
    for i in self.charlist_correctinput:
      print(i,end=' ')
  def key(self):
    key = input("\n\n\nDigite uma letra: ")
    return key
  #Método para checar o status do game e imprimir o jogo na tela
  def print_game_status(self):
    print(board[self.status])
    self.hide_word()
    self.wrong_chars()
    self.correct_chars()
```

```python
def randWord():
  with open("/content/palavras.txt", "rt") as f:
    bank = f.readlines()
  return bank[random.randint(0,len(bank)-1)].strip()
```

```python
def main():
  #Objeto
  game = Hangman(randWord())
  #Enquanto o jogo não tiver terminado, print de status, solicita uma letra e faz a leitura do caracter
  #Verifica o status do jogo
  while(game.hangman_over() == False):
    system('clear')
    game.print_game_status()
    game.guess(letra=game.key())
  game.print_game_status()
  #De acordo com o status, imprime mensagem na tela para o usuário
  if game.hangman_won():
    print('\n\nParabéns! Você venceu!!')
  else:
    print('\n\nGame Over! Você perdeu!!')
    print('\nA palavra era ' + game.word)
  print('\nFoi bom jogar com você! Agora vá estudar!\n')
```

```python
#Executa o programa
if __name__ == "__main__":
  main()
```
